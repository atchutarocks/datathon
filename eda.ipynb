```python
import pandas as pd

# Path to the JSONL file
jsonl_file_path = '/Users/acram/Downloads/public_dat/attrebute_train.solution'

# Read the JSONL file into a DataFrame
df = pd.read_json(jsonl_file_path, lines=True)

# Display the DataFrame
df.head()
```

```python
unique_brands = df['details_Brand'].unique()
print("\nUnique Brands:")
print(unique_brands)
print("Number of unique_brands in training data is:",len(unique_brands))
```

```python
unique_categories = {
    'L0_category': df['L0_category'].unique(),
    'L1_category': df['L1_category'].unique(),
    'L2_category': df['L2_category'].unique(),
    'L3_category': df['L3_category'].unique(),
    'L4_category': df['L4_category'].unique()
}

print("\nUnique Categories:")
for level, categories in unique_categories.items():
    print(f"\n{level}:")
    print(categories)
    print("Number of unique categories is:",len(categories))
```
